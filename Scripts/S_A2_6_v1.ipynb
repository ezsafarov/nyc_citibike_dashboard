```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# 1. Load the data
df = pd.read_csv("/Users/emilsafarov/Library/CloudStorage/OneDrive-Personal/CF/CF_S2/nyc_citibike_dashboard_old/Data/Output/citibike_weather_merged_2022.csv")

# 2. Group by existing 'date' column (already cleaned) to count trips
daily_trips = df.groupby("date")["ride_id"].count().reset_index(name="trip_count")

# 3. Use existing avg_temp column (already aligned by date)
daily_temp = df.groupby("date")["avg_temp"].mean().reset_index()

# 4. Merge into one DataFrame
df_daily = pd.merge(daily_trips, daily_temp, on="date")

# 5. Check results
print("✅ trip_count stats:")
print(df_daily["trip_count"].describe())
```

```python
# Create dual-axis plot
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Red line: Daily bike trips (right y-axis)
fig.add_trace(
    go.Scatter(
        x=df_daily["date"],
        y=df_daily["trip_count"],
        name="Daily Bike Trips",
        line=dict(color="red", width=2)
    ),
    secondary_y=True
)

# Blue line: Avg temperature (left y-axis)
fig.add_trace(
    go.Scatter(
        x=df_daily["date"],
        y=df_daily["avg_temp"],
        name="Avg Temperature (°C)",
        line=dict(color="blue", width=2)
    ),
    secondary_y=False
)

# Layout adjustments
fig.update_layout(
    title="Daily Bike Trip Count vs. Temperature (2022)",
    xaxis_title="Date",
    height=600,
    font=dict(size=14),
    plot_bgcolor="white",
    paper_bgcolor="white",
    legend=dict(orientation="h", y=-0.2)
)

fig.update_yaxes(title_text="Avg Temperature (°C)", secondary_y=False)
fig.update_yaxes(title_text="Trip Count", secondary_y=True)

# Show chart
fig.show()
```

```python
# Bike type and user role breakdown - output of the second task

# Count values
rideable_counts = df["rideable_type"].value_counts()
user_counts = df["member_casual"].value_counts()

# Create 1 row, 2 columns of subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Bar chart: Rideable type
ax1.bar(rideable_counts.index, rideable_counts.values, color="steelblue", edgecolor="black")
ax1.set_title("Rideable Type Distribution")
ax1.set_xlabel("Rideable Type")
ax1.set_ylabel("Count")

# Pie chart: Member vs Casual
ax2.pie(user_counts.values, labels=user_counts.index, autopct="%1.1f%%", startangle=90, colors=["orange", "lightgreen"])
ax2.set_title("User Type")

# Add overall title and layout
plt.suptitle("Bike Type and User Role Breakdown")
plt.tight_layout()
plt.show()
```

```python
# Top 20 Most Frequent Start Stations — Horizontal Bar Chart

# Set Seaborn theme and color palette for all plots
sns.set_theme(style="whitegrid", palette="Blues")

# Count number of rides per start station
df['value'] = 1  # helper column for counting
station_counts = df.groupby('start_station_name')['value'].count().reset_index()

# Get the top 20 most frequented start stations
top20 = station_counts.nlargest(20, 'value')

# Plot horizontal bar chart using a blue palette with 20 shades
plt.figure(figsize=(10, 6))
with sns.color_palette("Blues_r", n_colors=20):
    sns.barplot(data=top20, x='value', y='start_station_name')
    plt.title("Top 20 Most Frequent Start Stations")
    plt.xlabel("Number of Rides")
    plt.ylabel("Start Station")
    plt.tight_layout()
    plt.show()
```

```python
# Trip Duration by Rider Type — Box Plot (Filtered for 1–120 Minutes)
```

```python
# Convert datetime columns
df['started_at'] = pd.to_datetime(df['started_at'])
df['ended_at'] = pd.to_datetime(df['ended_at'])

# Calculate trip duration in minutes
df['trip_duration_min'] = (df['ended_at'] - df['started_at']).dt.total_seconds() / 60

# Filter trips: keep only durations between 1 and 120 minutes
df_filtered = df[(df['trip_duration_min'] > 0) & (df['trip_duration_min'] <= 120)]

# Plot box plot comparing trip durations by rider type
plt.figure(figsize=(8, 5))
sns.boxplot(
    x='member_casual',
    y='trip_duration_min',
    data=df_filtered,
    palette={'member': 'royalblue', 'casual': 'lightcoral'}
)
plt.title("Trip Duration by Rider Type")
plt.xlabel("Rider Type")
plt.ylabel("Trip Duration (minutes)")
plt.tight_layout()
plt.show()
```

```python
# Trip Duration Distribution by Rider Type — Facet Histogram (1–120 Minutes)

# Create FacetGrid of trip duration by rider type
g = sns.FacetGrid(df_filtered, col='member_casual', height=5, aspect=1.2)
g.map(sns.histplot, 'trip_duration_min', bins=30, color='steelblue')

# Set x-axis ticks manually to show in minutes (e.g., every 10 min)
for ax in g.axes.flat:
    ax.set_xticks(range(0, 130, 10))  # From 0 to 120 minutes, every 10

# Add a common title and labels
g.fig.suptitle("Trip Duration Distribution by Rider Type", fontsize=16, y=1.05)
g.set_axis_labels("Trip Duration (minutes)", "Count")
plt.tight_layout()
plt.show()
```
