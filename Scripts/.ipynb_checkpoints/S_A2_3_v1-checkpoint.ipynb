### Step 3: daily trips

```python
daily_trips = df.groupby("date").size().reset_index(name="trip_count")
```

```python
daily_trips
```

```python
df_daily = df.groupby("date").agg({
    "avg_temp": "mean",     # or "first" if avg_temp is same per day
    "ride_id": "count"      # or any column to count rows
}).reset_index()

df_daily = df_daily.rename(columns={"ride_id": "trip_count"})
```

```python
df_daily
```

```python
### Step 4: Create Dual-Axis Line Plot
```

```python
# Set the date as index if not already
df_daily["date"] = pd.to_datetime(df_daily["date"])
df_daily.set_index("date", inplace=True)

# Create dual-axis plot
fig, ax1 = plt.subplots(figsize=(12, 5))

# Plot temperature on left y-axis
ax1.plot(df_daily.index, df_daily["avg_temp"], color="tab:blue", label="Avg Temp (C)")
ax1.set_xlabel("Date")
ax1.set_ylabel("Average Temperature (C)", color="tab:blue")
ax1.tick_params(axis='y', labelcolor="tab:blue")

# Create right y-axis for trip counts
ax2 = ax1.twinx()
ax2.plot(df_daily.index, df_daily["trip_count"], color="tab:orange", label="Trip Count")
ax2.set_ylabel("Trip Count", color="tab:orange")
ax2.tick_params(axis='y', labelcolor="tab:orange")

# Title and layout
plt.title("Daily Bike Trip Count vs. Temperature (2022)")
fig.tight_layout()
plt.show()
```

### Explanation: Dual-Axis Plot Using Matplotlib

I used Matplotlib's object-oriented paradigm to create a dual-axis line plot in this chart. This approach gives more control over the figure and axes.

- I first created a figure and the main axis (`ax1`) using `plt.subplots()`.
- On `ax1`, I plotted the daily average temperature using the left y-axis.
- Then, I used `ax1.twinx()` to create a secondary y-axis (`ax2`) sharing the same x-axis.
- On `ax2`, I plotted the daily bike trip counts.
- Each y-axis is styled with its own color and label, and the chart is made tight with `fig.tight_layout()`.

This method allows for a clear visual comparison of two variables (temperature and ride count) on different scales over time.

### Bonus part

```python
# Convert to datetime if not already
df["started_at"] = pd.to_datetime(df["started_at"])
df["ended_at"] = pd.to_datetime(df["ended_at"])

# Calculate trip duration in seconds
df["tripduration"] = (df["ended_at"] - df["started_at"]).dt.total_seconds()
```

```python
df.head()
```

```python
# Filter durations  1 hour
x = df["tripduration"]
x = x[x <= 3600]

# Compute histogram bins
counts, bins = np.histogram(x, bins=50)
bin_centers = 0.5 * (bins[1:] + bins[:-1])

# Plot histogram + curve
plt.figure(figsize=(10, 5))
plt.hist(x, bins=bins, color='skyblue', edgecolor='black', alpha=0.6, label='Trip Duration Histogram')
plt.plot(bin_centers, counts, color='darkblue', label='Step Curve')
plt.title("Trip Duration Distribution (060 min)")
plt.xlabel("Trip Duration (seconds)")
plt.ylabel("Number of Trips")
plt.grid(True, axis="y", alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

#### Figure with Two Subplots (OO Approach)

This figure uses Matplotlib's object-oriented (OO) interface to display two separate plots side by side:

- A **bar chart** showing the count of each `rideable_type`
- A **pie chart** illustrating the distribution of `member_casual` user types

We use `plt.subplots()` to create two axes (`ax1`, `ax2`) within the same figure  not a dual axis, but two independent plots.

```python
df["rideable_type"].unique()
```

```python
df["member_casual"].unique()
```

```python
df["rideable_type"].value_counts()
```

```python
df["member_casual"].value_counts()
```

```python
# Count values
rideable_counts = df["rideable_type"].value_counts()
user_counts = df["member_casual"].value_counts()

# Create 1 row, 2 columns of subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Bar chart: Rideable type
ax1.bar(rideable_counts.index, rideable_counts.values, color="steelblue", edgecolor="black")
ax1.set_title("Rideable Type Distribution")
ax1.set_xlabel("Rideable Type")
ax1.set_ylabel("Count")

# Pie chart: Member vs Casual
ax2.pie(user_counts.values, labels=user_counts.index, autopct="%1.1f%%", startangle=90, colors=["orange", "lightgreen"])
ax2.set_title("User Type (Member vs Casual)")

# Add overall title and layout
plt.suptitle("Bike Type and User Role Breakdown")
plt.tight_layout()
plt.show()
```
